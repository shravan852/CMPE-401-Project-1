```python
!pip install ultralytics -q
from google.colab import drive
drive.mount('/content/drive')

import os, yaml, time
from ultralytics import YOLO

DRIVE = '/content/drive/MyDrive'
RESULTS_DIR = f'{DRIVE}/yolo_results'
yaml_path = f'{DRIVE}/visdrone_yolo/visdrone.yaml'

# Recreate yaml if needed
if not os.path.exists(yaml_path):
    CLASS_NAMES = ['pedestrian','people','bicycle','car','van','truck',
                   'tricycle','awning-tricycle','bus','motor']
    os.makedirs(f'{DRIVE}/visdrone_yolo', exist_ok=True)
    with open(yaml_path, 'w') as f:
        yaml.dump({'path': f'{DRIVE}/visdrone_yolo', 'train': 'images/train',
                   'val': 'images/val', 'nc': 10, 'names': CLASS_NAMES}, f)

model = YOLO('yolo11m.pt')
model.train(
    data=yaml_path,
    epochs=50,
    imgsz=640,
    batch=16,
    lr0=0.01,
    optimizer='SGD',
    project=RESULTS_DIR,
    name='exp_yolo11m_final',
    exist_ok=True,
    patience=15,
    save=True,
    plots=True,
)
print(f'Done! Weights at: {RESULTS_DIR}/exp_yolo11m_final/weights/best.pt')
```
